```python
import pandas as pd
import sqlite3
con = sqlite3.connect('../assignment.db')
cursor = con.cursor()

sql = "SELECT n_title,id FROM IMDB;"  # Select all entries from contacts table
try:
    cursor.execute(sql)
    con.commit()
    print("Entry successfully retrieved.")
except Exception as e:
    print("Exception occurred.",e)

result = cursor.fetchall()

#print(type(result))
#print(result)

print('\n')
con.close()
df = pd.DataFrame(result, columns=['title','id'])
df
```

```python
import requests
import time

api_search1 = "https://imdb-api.com/en/API/Ratings/k_tj3n43f9/"
api_search2 = "https://imdb-api.com/en/API/Ratings/k_q67cbnpo/"
api_search =" https://imdb-api.com/en/API/Title/"

plot = []
images= []

i_title=[]
i_id=[]
flag=0
bp=[]
for i in df['id']:
    if flag%2 == 0:
        r=requests.get(api_search+"k_tj3n43f9/"+i+'/Images')
    else:
        r=requests.get(api_search+"k_q67cbnpo/"+i+'/Images')
    
    content = r.json()
    
    print(content['title'])
    
    if df.loc[flag,'title'] == content['title']:
        print('No.',flag,' match')
        print(content['plot'])
        plot.append(content['plot'])
        
        imgL = content['images']['items']
        img = []
        for i in range(4):
            #print (imgL[i]['image'])
            img.append(imgL[i]['image'])
        
        print(img)
        images.append(img)
        
        i_title.append(content['title'])
        i_id.append(content['id'])

        
    else:
        print('No.',flag,' dont`t match, recorded')
        bp.append(flag)
        plot.append('erro')
        images.append('erro')
        
        i_title.append('erro')
        i_id.append('erro')

    flag = flag+1
    time.sleep(1)

print(bp)
```
