```python
import pandas as pd 
import sqlite3 

con = sqlite3.connect('assignment.db') 
cursor = con.cursor()   
sql ="SELECT DISTINCT title_Chinese,n_title,year,id,image FROM douban INNER JOIN IMDB ON douban.title_English=IMDB.title "

try:
    cursor.execute(sql)
    con.commit()
    print("Entry successfully retrieved.")
except Exception as e:
    print("Exception occurred.",e)
# Display results
result = cursor.fetchall()

con.close()
df = pd.DataFrame(result, columns=[ 
'title','n_title','year','id','image'
  ])
df.to_csv('base_info.csv',mode='w',index=False,sep="\t")#部分电影名称包含逗号，所以使用tab作为分隔符
df
```

```python
################################################################################################################

# EXTERNAL MODULES TO BE USED

################################################################################################################

from flask import Flask, render_template, redirect, url_for
import pandas as pd

app = Flask(__name__)
    
################################################################################################################

# WEB ROUTES FOR CONTROLLING ACCESS TO TEMPLATE VIEWS

################################################################################################################

@app.route("/")
def index():
    return render_template('index.html')


@app.route("/movies")
def movies():
    df = pd.read_csv("base_info.csv",delimiter="\t")
    list = df.to_dict('records')
    return render_template('movies.html', entries = list)


if __name__ == '__main__':
    app.run()
```
