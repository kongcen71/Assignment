```python
import pandas as pd
import sqlite3
con = sqlite3.connect('assignment.db')
cursor = con.cursor()

sql = "SELECT title_English,year FROM douban;"  # Select all entries from contacts table
try:
    cursor.execute(sql)
    con.commit()
    print("Entry successfully retrieved.")
except Exception as e:
    print("Exception occurred.",e)

result = cursor.fetchall()

print(type(result))
print(result)

print('\n')
con.close()
df = pd.DataFrame(result, columns=['title','year'])
df
```

```python
import requests
import time

api_search1 = "https://imdb-api.com/en/API/SearchMovie/k_tj3n43f9/"
api_search2 = "https://imdb-api.com/en/API/SearchMovie/k_q67cbnpo/"

mid = []
image= []
n_title=[]
flag=0
bp=[]
for i in df['title']:
    if flag%2 == 0:
        r=requests.get(api_search1+i)
    else:
        r=requests.get(api_search2+i)
    
    content = r.json()

    content = content["results"]

    #print(type(content))
    try:    
        info = content[0]

        print(info['title'],info['id'],info['image'],(info['description']))

        #print(df.loc[flag,'year'],type(df.loc[flag,'year']))
        #print(int(info['description'][1:5]),type(int(info['description'][1:5])))
        try:
            if df.loc[flag,'year'] == int(info['description'][1:5]):
                print('No.',flag,' match')
            else:
                print('Error, years don`t match')
                print('No.',flag,' dont`t match, recorded')
                bp.append(flag)
        except ValueError as v:
            print('Error, can`t` find years')
            print('No.',flag,' dont`t match, recorded')
            bp.append(flag)

        mid.append(info['id'])
        image.append(info['image'])
        n_title.append(info['title'])
     
    except IndexError as ide:
        print('Error, no feedback')
        print('No.',flag,' dont`t match, recorded')
        bp.append(flag)
        mid.append('erro')
        image.append('erro')
        n_title.append('erro')

    flag = flag+1
    time.sleep(1)

print(mid)
print(image)
print(bp)
```
