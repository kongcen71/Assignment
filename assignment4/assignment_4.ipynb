```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.douban.com/people/texieshudian/subject_doulists/movie"

fake_headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.79 Safari/537.36'
}
response = requests.get(url, headers=fake_headers)  # 请求参数里面把假的请求header加上
# 保存网页到本地
file_obj = open('douban.html', 'w')  # 以写模式打开名叫 douban.html的文件

#file_obj = open('douban.html','w', encoding="utf-8")  # 以写模式打开名叫 douban.html的文件，指定编码为utf-8
file_obj.write(response.content.decode('utf-8'))  # 把响应的html内容
file_obj.close()  # 关闭文件，结束写入
```

```python
from bs4 import BeautifulSoup
src = open("douban.html")
document = BeautifulSoup(src, 'lxml')
src.close()
#print(document)
ul = document.find("ul",attrs={"class":"doulist-list"})
print(ul)
```

```python
li = ul.find_all("h3")
print(li)
```

```python
link=[]
genre= []
#print(li[-1:-9])
for i in li[-9:-1]:
    url = i.find("a")
    c=url.get_text().strip()
    #print(c)
    #print(url['href'])
    link.append(url['href'])
    g=c[c.find('之')+1:]
    #g.replace('系列',' ')
    genre.append(g.strip("系列"))
print(link)
print(genre)
```

```python
import requests
import csv
import pandas as pd
import os
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver

fake_headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.79 Safari/537.36'
}

header = ['title','rate','director','country','year','genre']

data=[]
i=0
for web in link:
    page = requests.get(web, headers=fake_headers)
    soup = BeautifulSoup(page.content,'html.parser')
    
    for item in soup.find_all("div", attrs={"class":"doulist-item"}):
        if item.find('span', class_='rating_nums') is None:
            continue

        title  = item.find("div", attrs={"class":"title"}).a.get_text().strip()
        print(title)

        rate = item.find('span', class_='rating_nums').get_text().strip()
        print(rate)

        abstract = item.find("div", attrs={"class": "abstract"}).get_text().strip()
        #print(abstract)
        director = abstract[abstract.find('导演'):abstract.find('主演')]
        country = abstract[abstract.find('制片国家/地区: '):abstract.find('年份')]
        year =  abstract[abstract.find('年份'):]
        print(director,country,year)
        #print(len(director),len(country),len(year))
        #print(director.strip(),country.strip(),year.strip())
        #print(len(director),len(country),len(year))

        data.append((title,rate,director.strip("导演:"),country.strip("制片国家/地区:"),year.strip("年份"),genre[i]))
        #data.append((title, r, abstract))
    i=i+1

df = pd.DataFrame(data,columns=header)
df.to_csv('douban.csv', sep='\t',encoding='utf-8')#爬中文网站的时候一定要加编码
print(df)
```
